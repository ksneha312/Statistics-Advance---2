**Question1: Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis testing?**

**Ans:-**
**Definition of the Z-Statistic:-**The z-statistic (or z-score) is a measure that describes how many standard deviations a data point (or sample mean) is away from the population mean. It is commonly used in statistics to standardize scores, making them comparable across different datasets. The z-statistic is calculated using the following

**Relationship to the Standard Normal Distribution:-**
The z-statistic follows the standard normal distribution if the underlying data is normally distributed or if the sample size is sufficiently large (by the Central Limit Theorem). The standard normal distribution has a mean of 0 and a standard deviation of 1.

In this context, the z-statistic is used to convert a data point or sample statistic to a form where we can compare it to a standard normal distribution. This makes it easier to perform statistical inference, such as hypothesis testing or constructing confidence intervals.

**Z-Statistic in Hypothesis Testing:-**

The z-statistic is used in hypothesis testing to determine whether the observed data supports the null hypothesis. Here's how the z-statistic is typically applied in hypothesis testing:

1.State the hypotheses (null and alternative).

2.Calculate the z-statistic based on sample data.

3.Find the critical value or p-value from the standard normal distribution.

4.Make a decision to either reject or fail to reject the null hypothesis.

**Hypothesis Testing Example in Python:-**

Let’s go through an example of using the z-statistic in hypothesis testing, assuming a sample mean and a known population mean.

Scenario:

We want to test if the average score of a group of students differs from 80. The population standard deviation is known to be 10, and we have a sample of 25 students with a sample mean of 82.

Steps:

Null hypothesis (H0): The mean score is 80 (μ=80).
Alternative hypothesis (H1): The mean score is not equal to 80 (μ=80).

```python
import numpy as np
import scipy.stats as stats

# Given data
sample_mean = 82  # Sample mean
population_mean = 80  # Population mean (under H0)
population_std = 10  # Population standard deviation
sample_size = 25  # Sample size

# Step 1: Calculate the standard error
standard_error = population_std / np.sqrt(sample_size)

# Step 2: Calculate the z-statistic
z_statistic = (sample_mean - population_mean) / standard_error

# Step 3: Find the critical z-value for a two-tailed test at alpha = 0.05
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test

# Step 4: Calculate the p-value for the observed z-statistic
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed p-value

# Output results
print(f"Z-statistic: {z_statistic:.3f}")
print(f"Critical z-value: ±{critical_value:.3f}")
print(f"P-value: {p_value:.3f}")

# Step 5: Make a decision based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis (there is a significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

**Question2 : What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is very small (e.g., 0.01)?**

**Ans:-**
In Python, you can calculate the p-value for hypothesis testing using statistical libraries like SciPy, which provides functions for various statistical tests. Here's a brief explanation and how you can implement hypothesis testing and interpret the p-value in Python.

W a p-value in Hypothesis Testing
The p-value is a measure used in hypothesis testing to help you decide whether to reject the null hypothesis (H₀).
It tells you the probability of observing the data you have (or something more extreme) under the assumption that the null hypothesis is true.
If the p-value is small (e.g., below 0.05), it suggests that the observed data is unlikely to have occurred under the null hypothesis, so you reject the null hypothesis.
How is it Used?
Null Hypothesis (H₀): Assumes no effect or no difference.
Alternative Hypothesis (H₁ or Ha): Assumes there is an effect or a difference.
Significance Level (α): A threshold for deciding whether to reject H₀, often set at 0.05.
If p-value < α, you reject the null hypothesis.
If p-value ≥ α, you fail to reject the null hypothesis.


```python
import numpy as np
from scipy import stats

# Sample data (e.g., heights of 10 people)
sample_data = np.array([66, 68, 70, 64, 65, 67, 69, 71, 63, 66])

# Null hypothesis: The mean height is 65 inches.
# We will use a one-sample t-test.

# Perform the t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, 65)

# Print the result
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpreting the p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is significant evidence that the mean height is not 65 inches.")
else:
    print("Fail to reject the null hypothesis: There is insufficient evidence to suggest that the mean height is not 65 inches.")
```

**Question3:Compare and contrast the binomial and Bernoulli distributions.**

**Ans:-**The Binomial and Bernoulli distributions are both discrete probability distributions related to binary outcomes, but they are used in different scenarios.

**Key Differences:**

**Bernoulli Distribution:**

Definition: The Bernoulli distribution is a special case of the binomial distribution where the number of trials is 1.

**Parameters:**

p: The probability of success in a single trial.

Range: Takes only two possible values: 0 (failure) or 1 (success).

Application: Models a single trial with two possible outcomes, such as flipping a coin once or testing whether a customer buys a product in one attempt.

**Binomial Distribution:**

Definition: The Binomial distribution generalizes the Bernoulli distribution to multiple trials. It represents the number of successes in a fixed number of independent Bernoulli trials.

**Parameters:**

n: The number of trials.

p: The probability of success in each trial.

Range: Takes integer values from 0 to n, representing the number of successes in n trials.

Application: Models scenarios where there are multiple independent trials, each with two possible outcomes (e.g., flipping a coin multiple times or testing a batch of products for defects).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom

# Parameters
p = 0.5  # Probability of success
n = 10   # Number of trials for binomial distribution

# Bernoulli distribution (1 trial)
bernoulli_rv = bernoulli(p)
bernoulli_values = np.array([0, 1])
bernoulli_probs = bernoulli_rv.pmf(bernoulli_values)

# Binomial distribution (n trials)
binomial_rv = binom(n, p)
binomial_values = np.arange(0, n+1)
binomial_probs = binomial_rv.pmf(binomial_values)

# Plot the distributions
plt.figure(figsize=(10, 6))

# Plot Bernoulli distribution
plt.subplot(1, 2, 1)
plt.bar(bernoulli_values, bernoulli_probs, color='blue', alpha=0.6, label="Bernoulli")
plt.title("Bernoulli Distribution")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks(bernoulli_values)

# Plot Binomial distribution
plt.subplot(1, 2, 2)
plt.bar(binomial_values, binomial_probs, color='green', alpha=0.6, label="Binomial")
plt.title("Binomial Distribution (n=10)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")

# Display the plots
plt.tight_layout()
plt.show()
```

3.Constant probability of success (p): The probability of success

5.Discrete outcomes: We are interested in counting the number of successes across the trials.

When all these conditions hold, the binomial distribution is appropriate to model the number of successes in n independent trials.

The Bernoulli Distribution is actually a special case of the Binomial Distribution:

The Bernoulli distribution models a single trial with two possible outcomes (success or failure).

The Binomial distribution is an extension of the Bernoulli distribution, where it models the number of successes across multiple independent Bernoulli trials.

**Question5: What are the key properties of the Poisson distribution, and when is it appropriate to use this distribution?**

**Ans:-**
Key Properties of the Poisson Distribution:
The Poisson Distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time or space, given that these events occur with a known constant mean rate and independently of the time since the last event. It is often used to model rare or random events.

Here are the key properties of the Poisson distribution:

Parameter (λ):

The Poisson distribution is parameterized by
λ (lambda), which represents the average rate of events occurring in a fixed interval of time or space. It is the expected number of events.
λ must be a non-negative real number.
Probability Mass Function (PMF): The probability of observing exactly
k events in a fixed interval is given by the formula:

P(X=k)= k!/λ k e −λ,k=0,1,2,…

Where:

k is the number of events (non-negative integer),

λ is the average rate of occurrence (mean),

e is Euler’s number (approximately 2.71828).

Mean (Expected Value):

The expected value of a Poisson-distributed random variable
X is:𝐸[𝑋] = λ

Variance:

The variance of a Poisson-distributed random variable is:
Var(X)=λ

Skewness:

The Poisson distribution is right-skewed when λ is small. As λ increases, the distribution approaches a normal distribution (because of the Central Limit Theorem).

Memorylessness:

The Poisson distribution is memoryless in the sense that the number of events occurring in a future interval is independent of the events that have occurred in the past interval.

Appropriate use of the Poisson Distribution
The Poisson distribution is appropriate in the following cases:

Rare Events:

When events are rare or infrequent over a fixed period of time or space, such as the number of accidents at an intersection in a day, or the number of phone calls at a call center in an hour.

Fixed Interval:

When the interval (time or space) over which events are observed is fixed and known. For example, a 1-hour period, a 1-day period, or a 1-mile stretch of road.

Events Occur Independently:

The occurrence of one event does not influence the occurrence of another event. For example, one customer arriving at a store does not affect the probability of another customer arriving.

Constant Rate:

The events occur at a constant average rate 𝜆 over time or space. For example, if a factory experiences an average of 3 machine failures per week, the number of failures follows a Poisson distribution with λ=3.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_value = 3  # Average number of events (rate)
size = 1000  # Number of random samples to generate

# Generate random samples from Poisson distribution
poisson_samples = poisson.rvs(mu=lambda_value, size=size)

# Calculate the probabilities for k events
k_values = np.arange(0, 10)
poisson_probs = poisson.pmf(k_values, mu=lambda_value)

# Plot the histogram of random samples
plt.figure(figsize=(10, 6))

# Plot Poisson distribution histogram
plt.subplot(1, 2, 1)
plt.hist(poisson_samples, bins=range(11), density=True, color='skyblue', alpha=0.6, edgecolor='black')
plt.title(f"Poisson Distribution with λ={lambda_value}")
plt.xlabel("Number of Events (k)")
plt.ylabel("Probability")
plt.xticks(np.arange(0, 11))

# Plot the theoretical Poisson probability mass function (PMF)
plt.subplot(1, 2, 2)
plt.bar(k_values, poisson_probs, color='salmon', alpha=0.6, edgecolor='black')
plt.title(f"Theoretical Poisson Distribution PMF with λ={lambda_value}")
plt.xlabel("Number of Events (k)")
plt.ylabel("Probability")
plt.xticks(k_values)

# Display the plots
plt.tight_layout()
plt.show()
```

**Question6: Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)?**

**Ans:-**
Here's a breakdown of the terms "probability distribution," "probability density function (PDF)," and how a PDF differs from a probability mass function (PMF):

**1. Probability Distribution**

A probability distribution defines how the probabilities are assigned to each possible outcome of a random experiment. It can describe either discrete or continuous random variables. The distribution gives the probabilities of all possible values that a random variable can take.

**2. Probability Density Function (PDF)**

The probability density function (PDF) is used to describe the probability distribution of continuous random variables. The PDF gives the likelihood of a random variable falling within a particular range of values, but not the exact probability of a single value (since the probability for a single point in a continuous distribution is always zero).

The integral (area under the curve) of the PDF over a given range gives the probability that the random variable falls within that range.

**3. Probability Mass Function (PMF)**

A probability mass function (PMF) is used for discrete random variables. The PMF gives the probability that a discrete random variable is exactly equal to some value. Unlike the PDF, which deals with continuous distributions, the PMF sums the probabilities over a set of discrete values.

The sum of the probabilities for all possible outcomes must equal 1.
Example: Rolling a fair die (values from 1 to 6) or flipping a coin (values of heads or tails).

**Key Differences Between PDF and PMF**

**Nature of Random Variable:**

PDF applies to continuous random variables (e.g., real numbers).

PMF applies to discrete random variables (e.g., integers or finite sets).

**Probabilities:**

For a PDF, the probability of a specific outcome is always zero. Instead, you compute the probability over intervals.

For a PMF, the probability is directly associated with specific values.

**Summation vs. Integration:**

In PMF, the sum of probabilities over all possible outcomes is 1.

In PDF, the total probability is computed by integrating the function over the entire range.

```python
#Example: PMF for a Discrete Random Variable (e.g., Rolling a Die)
import numpy as np
import matplotlib.pyplot as plt

# Discrete random variable - Rolling a die (PMF)
outcomes = [1, 2, 3, 4, 5, 6]  # possible outcomes (1 to 6)
probabilities = [1/6] * 6  # probability for each outcome (fair die)

# Plot PMF
plt.bar(outcomes, probabilities, color='skyblue', edgecolor='black')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of Rolling a Die')
plt.show()
```

```python
#Example: PDF for a Continuous Random Variable (e.g., Normal Distribution)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Continuous random variable - Normal Distribution (PDF)
mean = 0  # mean of the distribution
std_dev = 1  # standard deviation of the distribution
x = np.linspace(-5, 5, 1000)  # range of x values
pdf_values = norm.pdf(x, mean, std_dev)  # PDF values

# Plot PDF
plt.plot(x, pdf_values, color='blue')
plt.fill_between(x, pdf_values, color='skyblue', alpha=0.5)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Normal Distribution')
plt.show()
```

**Question7: Explain the Central Limit Theorem (CLT) with example.**

**Ans:-**
**Central Limit Theorem (CLT)**

The Central Limit Theorem (CLT) is a fundamental theorem in statistics that describes how the distribution of the sample mean (or sum) of a large number of independent, identically distributed (i.i.d.) random variables approaches a normal distribution as the sample size increases, regardless of the original distribution of the data.

**Key Points of the CLT:**

Sample Means: When you take multiple samples from a population, calculate the mean for each sample, and then plot the distribution of those sample means, that distribution will approximate a normal distribution, as long as the sample size is large enough.

Sample Size: The sample size should be sufficiently large for the CLT to hold true (typically n>30 is considered large enough).

Original Distribution: The CLT works regardless of the shape of the original distribution (whether it's skewed, uniform, exponential, etc.).

**Example of Central Limit Theorem (CLT) in Python**

We'll simulate the CLT by:

Generating a random distribution (for example, an exponential distribution, which is not normal).

Drawing multiple samples from this distribution.

Calculating the sample means.

Plotting the distribution of those sample means to show how they approximate a normal distribution as the sample size increases.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 10000  # Large population
sample_size = 50  # Sample size for each sample
num_samples = 1000  # Number of samples to take

# Step 1: Generate a population with an exponential distribution
population = np.random.exponential(scale=1, size=population_size)

# Step 2: Take multiple samples and compute the sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Step 3: Plot the population distribution (Exponential Distribution)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='lightblue', edgecolor='black')
plt.title('Original Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 4: Plot the distribution of sample means (Central Limit Theorem in action)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Question8: Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be a pplied instead?**

**Ans:-**
**Z-Scores vs. T-Scores:**
Both z-scores and t-scores are used in statistical analyses to compare data points to the mean of a distribution. However, the key difference lies in the assumptions they make about the population and the sample size.

**1. Z-Score:**

A z-score is a measure that describes how many standard deviations a data point is from the mean of the population. It is used when:

The population standard deviation (σ) is known.

The sample size is large enough (typically n>30).

The data is assumed to be normally distributed.

**2. T-Score:**

A t-score is used when:

The population standard deviation is not known, and instead, we have to use the sample standard deviation.
The sample size is small (typically n<30).

The data is assumed to be normally distributed, or the sample size is small enough that we assume it might follow a t-distribution.

**When to Use Z-Scores vs. T-Scores:**

**Use Z-scores when:**

The population standard deviation is known, and

The sample size is large (n>30).

**Use T-scores when:**

The population standard deviation is unknown, and

The sample size is small (n<30).

```python
#1. Z-Score Example (Large Sample, Known Population Standard Deviation)

import numpy as np

# Population parameters
mu = 50  # Population mean
sigma = 10  # Population standard deviation

# Sample data
sample_data = np.array([55, 52, 47, 60, 49])

# Calculate z-score for each data point
z_scores = (sample_data - mu) / sigma

print("Z-Scores:", z_scores)
```

```python
#2. T-Score Example (Small Sample, Unknown Population Standard Deviation)

import numpy as np
from scipy import stats

# Sample data
sample_data = np.array([55, 52, 47, 60, 49])

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Use sample standard deviation
sample_size = len(sample_data)

# Calculate t-score for each data point
t_scores = (sample_data - sample_mean) / (sample_std / np.sqrt(sample_size))

print("T-Scores:", t_scores)
```

**Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?**

 **Task: Write Python code to calculate the z-score and p-value for the given data.**

  **Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.**

```python
from scipy.stats import norm

# Given values
sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 25
alpha = 0.05  # Significance level

# Step 1: Calculate the z-score
z_score = (sample_mean - population_mean) / (population_std / (sample_size ** 0.5))

# Step 2: Calculate the p-value (two-tailed test)
# P-value for two-tailed test: 2 * (1 - cdf(z_score)) because we are looking at both tails
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 3: Hypothesis testing
if p_value <= alpha:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

# Output the results
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
print(result)
```

**Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python. Generate 1,000 samples and plot the distribution. What is the expected mean and variance?**

**Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.**

**Objective: Understand the properties of a binomial distribution and verify them through simulation.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # Number of trials
p = 0.6     # Probability of success
size = 1000 # Number of samples

# Step 1: Simulate the binomial distribution
binomial_samples = np.random.binomial(n, p, size)

# Step 2: Plot the distribution (histogram)
plt.figure(figsize=(10, 6))
plt.hist(binomial_samples, bins=np.arange(0, n+2)-0.5, edgecolor='black', density=True, alpha=0.7)
plt.title(f"Binomial Distribution (n={n}, p={p}) - 1000 Samples")
plt.xlabel("Number of Successes")
plt.ylabel("Probability Density")
plt.xticks(np.arange(0, n+1))
plt.grid(True)
plt.show()

# Step 3: Calculate the mean and variance
mean = np.mean(binomial_samples)
variance = np.var(binomial_samples)

# Theoretical mean and variance
theoretical_mean = n * p
theoretical_variance = n * p * (1 - p)

# Output the results
print(f"Simulated Mean: {mean:.4f}")
print(f"Simulated Variance: {variance:.4f}")
print(f"Theoretical Mean: {theoretical_mean}")
print(f"Theoretical Variance: {theoretical_variance}")
```
